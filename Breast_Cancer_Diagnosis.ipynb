## **Checking accuracy for Breast Cancer Diagnosis Classifier with KNN, Logistic Regression, RF and SVM and comparing accuracy with and without Hyper-parameter Tuning and Feature Engineering**

```python
#importing  libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the file
df = pd.read_csv(r'C:\Users\Priti\Desktop\corpnce_data_science\kaggle proj\breast_cancer_data.csv')
print(df.shape)
print(df.columns)
print(df.columns.size)
```

```python
df.head()
```

```python
df.info() # no null values
```

```python
df.describe()
```

```python
#deleting 2 columns which are not usefull
df.drop(['id','Unnamed: 32'],axis=1,inplace=True)
```

```python
#counting diagnosis variable
df.diagnosis.value_counts()
```

```python
#diagnosis is the label data, convert it to numeric
#replacing M and B with 1 and 0 respectively
df.diagnosis.replace(('M','B'),(1,0),inplace=True)
```

```python
df.head()
```

**Creating a model with no hyper parameter tuning and no feature engineering**


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
```

```python
#separating feature data and labels

X = df.iloc[:,1:]
y= df.diagnosis
print(X.shape)
print(y.shape)
```

```python
#Splitting data into train and test
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.2,random_state=42)
print(X_train.shape)
print(X_test.shape)
```

**Logistic regression without hyperparameter tuning and feature engineering**

```python

log = LogisticRegression()
log.fit(X_train,y_train)
pred_log = log.predict(X_test)
print('Accuracy score of logistic regression is : ', metrics.accuracy_score(pred_log,y_test),'\n')
print('confusion matrix:\n', metrics.confusion_matrix(pred_log,y_test))
```

**Random forest classifier without hyperparameter tuning and feature engineering**

```python
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
pred_rf = rf.predict(X_test)
print('Accuracy score of logistic regression is : ', metrics.accuracy_score(pred_rf,y_test),'\n')
print('confusion matrix:\n', metrics.confusion_matrix(pred_rf,y_test))
```

**SVM without hyperparameter tuning and feature engineering**

```python
svm = SVC()
svm.fit(X_train,y_train)
pred_svm = svm.predict(X_test)
print('Accuracy score of logistic regression is : ', metrics.accuracy_score(pred_svm,y_test),'\n')
print('confusion matrix:\n', metrics.confusion_matrix(pred_svm,y_test))
```

**KNN without hyperparameter tuning and feature engineering**

```python
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
pred_knn = knn.predict(X_test)
print('Accuracy score of logistic regression is : ', metrics.accuracy_score(pred_knn,y_test),'\n')
print('confusion matrix:\n', metrics.confusion_matrix(pred_knn,y_test))
```

**Create model by tuning hyper parameter tuning and feature engineering**

```python
#standerdising the data 
std = StandardScaler()
X_std = std.fit_transform(X)
X_std = pd.DataFrame(X_std)
X_std.describe()
```

```python
X_std_train,X_std_test,y_train,y_test = train_test_split(X_std,y,test_size=.2,random_state=42)
print(X_std_train.shape)
print(X_std_test.shape)
```

**Logistic Regression with hyper parameter tuning and feature engineering**


```python

log = LogisticRegression()
param_grid = {
    'C': [.01,.1,1]
}

grid = GridSearchCV(log, param_grid, cv = 7)
print(grid.fit(X_std_train, y_train))
```

```python
print(grid.best_params_)
```

```python
log = LogisticRegression(C=0.1)
log.fit(X_std_train,y_train)
pred_log = log.predict(X_std_test)
print('Accuracy score of logistic regression is : ', metrics.accuracy_score(pred_log,y_test),'\n')
print('confusion matrix:\n', metrics.confusion_matrix(pred_log,y_test))
```

**Random Forest with hyper parameter tuning and feature engineering**

```python
rf = RandomForestClassifier(random_state=42, n_jobs=-1)
param_grid ={'criterion':['gini','entropy'],
          'n_estimators':[10,30,50],
           'max_depth' :[2,3,4],
             'n_jobs':[-1]
          }

grid = GridSearchCV(rf, param_grid, cv = 7)
print(grid.fit(X_std_train, y_train))
```

```python
grid.best_params_
```

```python
rf = RandomForestClassifier(criterion='entropy',max_depth=4,n_estimators=10 ,random_state=42,n_jobs=-1)
rf.fit(X_std_train,y_train)
pred_rf = rf.predict(X_std_test)
print('Accuracy score of logistic regression is : ', metrics.accuracy_score(pred_rf,y_test),'\n')
print('confusion matrix:\n', metrics.confusion_matrix(pred_rf,y_test))
```

**SVM with hyper parameter tuning and featire engineering**

```python
svm = SVC()
param_grid ={'C':[.01,.1,1,10],
             'kernel':['linear','rbf','poly']
          }

grid = GridSearchCV(svm, param_grid, cv = 7)
print(grid.fit(X_std_train, y_train))
```

```python
grid.best_params_
```

```python
svm = SVC(C=.1,kernel='linear')
svm.fit(X_std_train,y_train)
pred_svm = svm.predict(X_std_test)
print('Accuracy score of logistic regression is : ', metrics.accuracy_score(pred_svm,y_test),'\n')
print('confusion matrix:\n', metrics.confusion_matrix(pred_svm,y_test))
```

**KNN with hyper parameter tuning and feature engineering**

```python
knn = KNeighborsClassifier(n_jobs=-1)
param_grid ={'n_neighbors':[5,6,7,8,9,10],
             'n_jobs':[-1]
          }

grid = GridSearchCV(knn, param_grid, cv = 7)
print(grid.fit(X_std_train, y_train))
```

```python
grid.best_params_
```

```python
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(X_std_train,y_train)
pred_knn = knn.predict(X_std_test)
print('Accuracy score of logistic regression is : ', metrics.accuracy_score(pred_knn,y_test),'\n')
print('confusion matrix:\n', metrics.confusion_matrix(pred_knn,y_test))
```
